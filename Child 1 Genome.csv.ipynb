## Visualize the number of SNPs per chromosome
Chromosomes with higher/lower SNP counts indicate regions of dense genetic information or gaps.

```python
# SNPs distribution
df['chromosome'].value_counts().plot(kind='bar')
plt.show()
```

## Identify the frequency of each allele (e.g., A, G, T, C) in the dataset
This chart shows which alleles are most common in the dataset


```python
from collections import Counter
import matplotlib.pyplot as plt

# Count alleles
alleles = ''.join(df['genotype'].dropna())
allele_count = Counter(alleles)

# Plot
plt.bar(allele_count.keys(), allele_count.values(), color='skyblue')
plt.xlabel('Alleles')
plt.ylabel('Frequency')
plt.title('Allele Frequency Distribution')
plt.show()


```

## Visualize the number of SNPs per chromosome
Chromosomes with higher/lower SNP counts indicate regions of dense genetic information or gaps.


```python

chrom_counts = df['chromosome'].value_counts()

# Bar plot for chromosome distribution
chrom_counts.plot(kind='bar', color='purple', alpha=0.7)
plt.xlabel('Chromosome')
plt.ylabel('Number of SNPs')
plt.title('SNP Distribution Across Chromosomes')
plt.show()
```

## Show proportion of SNPs associated with disease risk.
This chart highlights what percentage of the genome might contribute to disease susceptibility.


```python

# Filter disease-related SNPs
risk_snps = df[df['# rsid'] == df['genotype']]

# Create pie chart
labels = ['Risk SNPs', 'Non-Risk SNPs']
sizes = [len(risk_snps), len(df) - len(risk_snps)]
colors = ['red', 'green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Disease-Associated SNPs')
plt.show()
# This chart highlights what percentage of the genome might contribute to disease susceptibility.
```

## Visualize the frequency of each genotype (e.g., AA, AG, GG).
This can indicate whether homozygous or heterozygous genotypes are more prevalent.


```python

genotype_counts = df['genotype'].value_counts()

# Bar plot for genotype counts
genotype_counts.plot(kind='bar', color='orange', alpha=0.8)
plt.xlabel('Genotype')
plt.ylabel('Frequency')
plt.title('Genotype Frequency Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='position', y='chromosome', hue='# rsid', palette='cool', alpha=0.6)
plt.title('Chromosomal SNP Positions')
plt.xlabel('Position on Chromosome')
plt.ylabel('Chromosome')
plt.legend(title='Risk Allele')
plt.show()
```
